```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
titanic = sns.load_dataset('titanic')
```

# **Who Survived, Male or Female?**
The histogram depictes that though the number of Female passangers is low, percentage of survival is more by Females than Male passangers.
Highest of 20-40 years age group travelled on the Titanic which clearly shows from the graph.

```python
sns.set_theme(style="whitegrid")
sns.displot(data=titanic,x="age",col="sex",hue="survived",kde=True)
```

```python
sns.catplot(data=titanic,x="survived",y="age",kind="box",aspect=1.5)
```
